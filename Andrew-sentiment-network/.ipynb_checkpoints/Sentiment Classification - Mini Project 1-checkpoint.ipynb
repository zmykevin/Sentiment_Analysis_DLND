# Lesson: Curate a Dataset

```python
def pretty_print_review_and_label(i):
    print(labels[i] + "\t:\t" + reviews[i][:80] + "...")

g = open('reviews.txt','r') # What we know!
reviews = list(map(lambda x:x[:-1],g.readlines()))
g.close()

g = open('labels.txt','r') # What we WANT to know!
labels = list(map(lambda x:x[:-1].upper(),g.readlines()))
g.close()
```

# Lesson: Develop a Predictive Theory

```python
print("labels.txt \t : \t reviews.txt\n")
pretty_print_review_and_label(2137)
pretty_print_review_and_label(12816)
pretty_print_review_and_label(6267)
pretty_print_review_and_label(21934)
pretty_print_review_and_label(5297)
pretty_print_review_and_label(4998)
```

```python
#Import necessary libaries
from collections import Counter
import numpy as np
```

```python
#Define the counters for positive and negative words
positive_counter = Counter()
negative_counter = Counter()
total_counter = Counter()
```

```python
#Count the positive and negative words in the reiview
for i in range(len(reviews)):
    current_review_words = reviews[i].split()
    if(labels[i] == 'POSITIVE'):
        for word in current_review_words:
            positive_counter[word] += 1
            total_counter[word] += 1
    else:
        for word in current_review_words:
            negative_counter[word] += 1
            total_counter[word] += 1
```

```python
positive_counter.most_common()
```

```python
#Define a new concept called positive negative ratio
#Which models the frequency of a word appear in each category
pos_neg_ratio_counter = Counter()

for term,cnt in list(total_counter.most_common()):
    if(cnt > 100):
        pos_neg_ratio = positive_counter[term]/(negative_counter[term]+1)
        pos_neg_ratio_counter[term] = pos_neg_ratio
for word,ratio in pos_neg_ratio_counter.most_common():
    if(ratio>1):
        pos_neg_ratio_counter[word] = np.log(ratio)
    else:
        pos_neg_ratio_counter[word] = -np.log(1/(ratio+0.01))
```

```python
pos_neg_ratio_counter.most_common() #Most frequent words that appear in positive reviews
```

```python
list(reversed(pos_neg_ratio_counter.most_common()))[0:30] #Most frequent words that appear in negative reviews
```
